```python
# Selection of relevant columns
df = df[['SHEM',
'HACNASA_AZMAI_MMUZA',
 'SACHAR_SCHIR_MMUZA',
 'MEUCHADET',
 'MAKABI',
 'LEUMIT',
 'KLAKIT',
 'ZKN',
 'MVGR',
 'YLD',
 'TOSH',]]
df = df.drop(index=0)
df = df.reset_index(drop=True)
# Data is relatively clean and ready for processing and visualizations
# There is a spelling error here: not 'KLAKIT', it must been 'KLALIT'
df.rename(columns={'KLAKIT':'KLALIT'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# רשימת הקופות חולים
kupot = ['MEUCHADET', 'MAKABI', 'LEUMIT', 'KLALIT']

# הגדרת צבעים לכל קופת חולים
colors = {
    'MEUCHADET': 'blue',
    'MAKABI': 'orange',
    'LEUMIT': 'green',
    'KLALIT': 'red'
}

# ניקוי של כל העמודות המספריות (המרה לאינטגרלים אחרי הסרת רווחים)
for column in kupot:
    df[column] = df[column].str.replace(',', '')  # הסרת פסיקים אם ישנם
    df[column] = df[column].str.strip()  # הסרת רווחים מיותרים
    df[column] = pd.to_numeric(df[column], errors='coerce')  # המרת נתונים למספרים

# סכימה של כל החברים בכל קופת חולים לאחר ניקוי הנתונים
total_per_kupah = df[kupot].sum()

# יצירת פי פלוט לאחר ניקוי הנתונים עם צבעים מותאמים
plt.figure(figsize=(8, 8))
plt.pie(total_per_kupah, labels=total_per_kupah.index, autopct='%1.1f%%', startangle=140, colors=[colors[kupah] for kupah in total_per_kupah.index])
plt.title('Distribution of Health Insurance Popularity in Israel')
plt.show()
```

```python
# בחירת חמש הערים עם המספר הגבוה ביותר של חברים בכל קופת חולים
top_cities_meuchedet = df.nlargest(5, 'MEUCHADET')[['SHEM', 'MEUCHADET']].rename(columns={'SHEM': 'City', 'MEUCHADET': 'MEUCHADET'})
top_cities_makabi = df.nlargest(5, 'MAKABI')[['SHEM', 'MAKABI']].rename(columns={'SHEM': 'City', 'MAKABI': 'MAKABI'})
top_cities_leumit = df.nlargest(5, 'LEUMIT')[['SHEM', 'LEUMIT']].rename(columns={'SHEM': 'City', 'LEUMIT': 'LEUMIT'})
top_cities_klalit = df.nlargest(5, 'KLALIT')[['SHEM', 'KLALIT']].rename(columns={'SHEM': 'City', 'KLALIT': 'KLALIT'})

# איחוד כל התוצאות לטבלת DataFrame אחת
top_cities_per_kupah = pd.concat([top_cities_meuchedet, top_cities_makabi, top_cities_leumit, top_cities_klalit], axis=1)

# רשימת הקופות חולים
kupot = ['MEUCHADET', 'MAKABI', 'LEUMIT', 'KLALIT']

# הגדרת צבעים שונים לכל קופת חולים
colors = {
    'MEUCHADET': 'blue',
    'MAKABI': 'orange',
    'LEUMIT': 'green',
    'KLALIT': 'red'
}

# פונקציה לניקוי שם העיר
def clean_city_name(city_name):
    return city_name.strip()[::-1]  # הסרת רווחים והיפוך הטקסט
df['SHEM'] = df['SHEM'].apply(clean_city_name)

# יצירת גרף לכל קופת חולים
for kupah in kupot:
    # בחירת חמש הערים עם המספר הגבוה ביותר של חברים בקופה הנוכחית
    top_cities = df[['SHEM'] + kupot].nlargest(5, kupah)

    # הגדרת מיקום העמודות (לפי מספר הערים)
    x = np.arange(len(top_cities['SHEM']))  
    width = 0.2  # רוחב כל עמודה

    # יצירת גרף עמודות מרובות קבוצות
    plt.figure(figsize=(12, 8))
    for i, other_kupah in enumerate(kupot):
        plt.bar(x + i * width, top_cities[other_kupah], width, label=other_kupah, color=colors[other_kupah])
    
    plt.xlabel('City')
    plt.ylabel('Number of Members')
    plt.title(f'Comparison of Health Insurance Members in Top 5 Cities for {kupah}')
    plt.xticks(x + width * (len(kupot) - 1) / 2, top_cities['SHEM'])
    plt.legend(title='Kupat Holim')
    plt.show()
```

```python
# ניקוי של כל העמודות המספריות (המרה לאינטגרלים אחרי הסרת רווחים)
for column in ['HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA', 'MEUCHADET', 'MAKABI', 'LEUMIT', 'KLALIT', 'TOSH']:
    if df[column].dtype == 'object':  # ניקוי רק אם העמודה מכילה מחרוזות
        df[column] = df[column].str.replace(',', '')  # הסרת פסיקים אם ישנם
        df[column] = df[column].str.strip()  # הסרת רווחים מיותרים
        df[column] = pd.to_numeric(df[column], errors='coerce')  # המרת נתונים למספרים


# 1. השוואת שכר שכיר ועצמאי ב-5 הערים הכי גדולות
top_5_cities_large = df.nlargest(5, 'TOSH')[['SHEM', 'HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA']]
plt.figure(figsize=(12, 8))
bar_width = 0.35
index = np.arange(len(top_5_cities_large))

plt.bar(index - bar_width/2, top_5_cities_large['HACNASA_AZMAI_MMUZA'], bar_width, label='Self-Employed Income', color='blue')
plt.bar(index + bar_width/2, top_5_cities_large['SACHAR_SCHIR_MMUZA'], bar_width, label='Salaried Income', color='orange')

plt.xlabel('City')
plt.ylabel('Income')
plt.title('Comparison of Self-Employed and Salaried Income in Top 5 Largest Cities')
plt.xticks(index, top_5_cities_large['SHEM'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 2. השוואת שכר שכיר ועצמאי ב-5 הערים הכי קטנות
bottom_5_cities_small = df.nsmallest(5, 'TOSH')[['SHEM', 'HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA']]
plt.figure(figsize=(12, 8))
index = np.arange(len(bottom_5_cities_small))

plt.bar(index - bar_width/2, bottom_5_cities_small['HACNASA_AZMAI_MMUZA'], bar_width, label='Self-Employed Income', color='blue')
plt.bar(index + bar_width/2, bottom_5_cities_small['SACHAR_SCHIR_MMUZA'], bar_width, label='Salaried Income', color='orange')

plt.xlabel('City')
plt.ylabel('Income')
plt.title('Comparison of Self-Employed and Salaried Income in Top 5 Smallest Cities')
plt.xticks(index, bottom_5_cities_small['SHEM'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()


average_income = df[['HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA']].mean()

# יצירת גרף עם שני ברים
plt.figure(figsize=(8, 6))
average_income.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Income for Self-Employed vs Salaried Workers in Israel')
plt.ylabel('Average Income')
plt.xticks(ticks=[0, 1], labels=['Self-Employed Income', 'Salaried Income'], rotation=0)
plt.tight_layout()
plt.show()
```
