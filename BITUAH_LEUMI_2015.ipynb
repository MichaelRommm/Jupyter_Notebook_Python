```python
# Israel population data (December 2015) - government databases, Provided by National Insurance (Bituah Leumi)
# Data includes only cities with a population of up to 2000 people
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Path to XML file
xml_file = "allinternet.xml"
# Read the XML file into a DataFrame, searching for the required encoding for Hebrew, specifying the encoding
df = pd.read_xml(xml_file, encoding='windows-1255')
```

```python
# Understanding how exactly city names are written and wich are the unique values ​​here 
print(df['SHEM'].unique())
# Apparently spaces at the beginning and end of the name of cities for improving data processing and preventing errors
df[df['SHEM'] == ' בת ים '][['TOSH', 'SHEM']]
```

```python
# Selection of relevant columns
df = df[['SHEM',
'HACNASA_AZMAI_MMUZA',
 'SACHAR_SCHIR_MMUZA',
 'MEUCHADET',
 'MAKABI',
 'LEUMIT',
 'KLAKIT',
 'ZKN',
 'MVGR',
 'YLD',
 'TOSH',]]
df = df.drop(index=0)
df = df.reset_index(drop=True)
# Data is relatively clean and ready for processing and visualizations
# There is a spelling error here: not 'KLAKIT', it must been 'KLALIT'
df.rename(columns={'KLAKIT':'KLALIT'}, inplace=True)
```

```python
# List of health funds
kupot = ['MEUCHADET', 'MAKABI', 'LEUMIT', 'KLALIT']

# Define colors for each health fund
colors = {
    'MEUCHADET': 'blue',
    'MAKABI': 'orange',
    'LEUMIT': 'green',
    'KLALIT': 'red'
}

# Clean all numeric columns (convert to integers after removing spaces)
for column in kupot:
    df[column] = df[column].str.replace(',', '')  # Remove commas if present
    df[column] = df[column].str.strip()  # Remove extra spaces
    df[column] = pd.to_numeric(df[column], errors='coerce')  # Convert to numbers

# Sum of all members in each health fund after cleaning the data
total_per_kupah = df[kupot].sum()

# Create a pie chart with customized colors after cleaning the data
plt.figure(figsize=(8, 8))
plt.pie(total_per_kupah, labels=total_per_kupah.index, autopct='%1.1f%%', startangle=140, colors=[colors[kupah] for kupah in total_per_kupah.index])
plt.title('Distribution of Health Insurance Popularity in Israel')
plt.show()

# Select the top five cities with the highest number of members for each health fund
top_cities_meuchedet = df.nlargest(5, 'MEUCHADET')[['SHEM', 'MEUCHADET']].rename(columns={'SHEM': 'City', 'MEUCHADET': 'MEUCHADET'})
top_cities_makabi = df.nlargest(5, 'MAKABI')[['SHEM', 'MAKABI']].rename(columns={'SHEM': 'City', 'MAKABI': 'MAKABI'})
top_cities_leumit = df.nlargest(5, 'LEUMIT')[['SHEM', 'LEUMIT']].rename(columns={'SHEM': 'City', 'LEUMIT': 'LEUMIT'})
top_cities_klalit = df.nlargest(5, 'KLALIT')[['SHEM', 'KLALIT']].rename(columns={'SHEM': 'City', 'KLALIT': 'KLALIT'})

# Combine all results into one DataFrame
top_cities_per_kupah = pd.concat([top_cities_meuchedet, top_cities_makabi, top_cities_leumit, top_cities_klalit], axis=1)

# Function to clean city names
def clean_city_name(city_name):
    return city_name.strip()[::-1]  # Remove spaces and reverse the text
df['SHEM'] = df['SHEM'].apply(clean_city_name)

# Create a chart for each health fund
for kupah in kupot:
    # Select the top five cities with the highest number of members in the current fund
    top_cities = df[['SHEM'] + kupot].nlargest(5, kupah)

    # Define bar positions (according to the number of cities)
    x = np.arange(len(top_cities['SHEM']))  
    width = 0.2  # Width of each bar

    # Create a grouped bar chart
    plt.figure(figsize=(12, 8))
    for i, other_kupah in enumerate(kupot):
        plt.bar(x + i * width, top_cities[other_kupah], width, label=other_kupah, color=colors[other_kupah])
    
    plt.xlabel('City')
    plt.ylabel('Number of Members')
    plt.title(f'Comparison of Health Insurance Members in Top 5 Cities for {kupah}')
    plt.xticks(x + width * (len(kupot) - 1) / 2, top_cities['SHEM'])
    plt.legend(title='Health Fund')
    plt.show()
```

```python
# Clean all numerical columns (convert to integers after removing spaces)
for column in ['HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA', 'MEUCHADET', 'MAKABI', 'LEUMIT', 'KLALIT', 'TOSH']:
    if df[column].dtype == 'object':  # Clean only if the column contains strings
        df[column] = df[column].str.replace(',', '')  # Remove commas if any
        df[column] = df[column].str.strip()  # Remove extra spaces
        df[column] = pd.to_numeric(df[column], errors='coerce')  # Convert data to numbers

# 1. Comparison of Self-Employed and Salaried Income in the Top 5 Largest Cities
top_5_cities_large = df.nlargest(5, 'TOSH')[['SHEM', 'HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA']]
plt.figure(figsize=(12, 8))
bar_width = 0.35
index = np.arange(len(top_5_cities_large))

plt.bar(index - bar_width/2, top_5_cities_large['HACNASA_AZMAI_MMUZA'], bar_width, label='Self-Employed Income', color='blue')
plt.bar(index + bar_width/2, top_5_cities_large['SACHAR_SCHIR_MMUZA'], bar_width, label='Salaried Income', color='orange')

plt.xlabel('City')
plt.ylabel('Income')
plt.title('Comparison of Self-Employed and Salaried Income in Top 5 Largest Cities')
plt.xticks(index, top_5_cities_large['SHEM'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# 2. Comparison of Self-Employed and Salaried Income in the Top 5 Smallest Cities
bottom_5_cities_small = df.nsmallest(5, 'TOSH')[['SHEM', 'HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA']]
plt.figure(figsize=(12, 8))
index = np.arange(len(bottom_5_cities_small))

plt.bar(index - bar_width/2, bottom_5_cities_small['HACNASA_AZMAI_MMUZA'], bar_width, label='Self-Employed Income', color='blue')
plt.bar(index + bar_width/2, bottom_5_cities_small['SACHAR_SCHIR_MMUZA'], bar_width, label='Salaried Income', color='orange')

plt.xlabel('City')
plt.ylabel('Income')
plt.title('Comparison of Self-Employed and Salaried Income in Top 5 Smallest Cities')
plt.xticks(index, bottom_5_cities_small['SHEM'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Average income
average_income = df[['HACNASA_AZMAI_MMUZA', 'SACHAR_SCHIR_MMUZA']].mean()

# Create a bar chart with two bars
plt.figure(figsize=(8, 6))
average_income.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Income for Self-Employed vs Salaried Workers in Israel')
plt.ylabel('Average Income')
plt.xticks(ticks=[0, 1], labels=['Self-Employed Income', 'Salaried Income'], rotation=0)
plt.tight_layout()
plt.show()
```

```python
# List of categories and their corresponding labels
categories = ['YLD', 'MVGR', 'ZKN']
labels = {
    'YLD': 'Children',
    'MVGR': 'Adults',
    'ZKN': 'Seniors'
}

# Define colors for each label
colors = {
    'Children': 'blue',
    'Adults': 'orange',
    'Seniors': 'green'
}

# Clean the data by converting columns to strings, removing commas, and extra whitespace
for column in categories:
    df[column] = df[column].astype(str)  # Convert columns to string
    df[column] = df[column].str.replace(',', '')  # Remove commas
    df[column] = df[column].str.strip()  # Remove extra whitespace
    df[column] = pd.to_numeric(df[column], errors='coerce')  # Convert to numeric, coercing errors

# Add a 'Total Population' column
df['Total Population'] = df[categories].sum(axis=1)

# --- Pie Chart: Distribution of Population Types ---

# Sum of all categories after cleaning the data
total_per_category = df[categories].sum()

# Create a pie chart with customized colors and labels
plt.figure(figsize=(8, 8))
plt.pie(total_per_category, 
        labels=[labels[cat] for cat in total_per_category.index], 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=[colors[labels[cat]] for cat in total_per_category.index])
plt.title('Distribution of Population Types')
plt.show()

# --- Bar Plot: Top 5 Largest Cities ---

# Get the top 5 largest cities based on 'Total Population'
top_5_cities = df.nlargest(5, 'Total Population')

# Melt the DataFrame for easier plotting with seaborn
df_melted_top = top_5_cities.melt(id_vars=['SHEM'], value_vars=categories, var_name='Population Type', value_name='Count')

# Map category codes to labels for the legend
df_melted_top['Population Type'] = df_melted_top['Population Type'].map(labels)

# Create a bar plot for the top 5 cities
plt.figure(figsize=(12, 8))
sns.barplot(data=df_melted_top, x='SHEM', y='Count', hue='Population Type', palette=colors)
plt.title('Population Types in the 5 Largest Cities')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Population Type')
plt.show()

# --- Bar Plot: 5 Smallest Cities ---

# Get the 5 smallest cities based on 'Total Population'
smallest_5_cities = df.nsmallest(5, 'Total Population')

# Melt the DataFrame for easier plotting with seaborn
df_melted_smallest = smallest_5_cities.melt(id_vars=['SHEM'], value_vars=categories, var_name='Population Type', value_name='Count')

# Map category codes to labels for the legend
df_melted_smallest['Population Type'] = df_melted_smallest['Population Type'].map(labels)

# Create a bar plot for the 5 smallest cities
plt.figure(figsize=(12, 8))
sns.barplot(data=df_melted_smallest, x='SHEM', y='Count', hue='Population Type', palette=colors)
plt.title('Population Types in the 5 Smallest Cities')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Population Type')
plt.show()
```
